```python
import pandas as pd

# Cargamos el archivo y lo mostramos
archivo_excel = "/content/Servicios_Sociales.xlsx"
datos_excel = pd.read_excel(archivo_excel)
print(datos_excel)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Datos de ejemplo
datos = {'Sexo': 15, ' Centro': 8, 'Distrito Centro': 3, 'Proceso': 6,
         'Barrio': 7, 'Nacionalidad': 4, 'Tramo Edad': 2, 'Fecha Atención': 2}

# Creamos un objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(datos)

# Mostramos la nube de etiquetas
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
archivo_excel2 = "/content/Sub.xls"
datos_excel2 = pd.read_excel(archivo_excel2)
```

```python
import matplotlib.pyplot as plt

# Guardar las columnas en variables independientes
asociacion = datos_excel2["ASOCIACIÓN"]
subvencion_euros = datos_excel2["SUBVENCIÓN EN EUROS"]

# Definir iconos para representar unidades
icono = 'o'

# Crear el gráfico isotópico
plt.figure(figsize=(8, 4))
plt.scatter(asociacion, [1] * len(asociacion), s=subvencion_euros, marker=icono)

# Personalizar el gráfico
plt.title('Gráfico Isotópico de Sub. en euros por Asociación')
plt.xlabel('Asociación')
plt.xticks(rotation=90)  # Ajustamos el ángulo de rotación
plt.ylabel('Subvención')
plt.yticks([])
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos ficticios
x = np.arange(1, 11)
y1 = np.array([0, 5, 10, 5, 10, 15, 10, 5, 0, 0])
y2 = np.array([0, 0, 0, 5, 10, 5, 10, 15, 10, 5])
y3 = np.array([5, 10, 15, 10, 5, 0, 0, 0, 0, 0])

# Crear un gráfico de flujo
fig, ax = plt.subplots()
ax.fill_between(x, 0, y1, label='Categoría', color='blue', alpha=0.7)
ax.fill_between(x, y1, y1 + y2, label='Categoría 2', color='green', alpha=0.7)
ax.fill_between(x, y1 + y2, y1 + y2 + y3, label='Categoría 3', color='orange', alpha=0.7)

# Personalizar el gráfico
ax.set_xlabel('Tiempo')
ax.set_ylabel('Valor')
ax.set_title('Gráfico de Flujo (Streamgraph)')
ax.legend(loc='upper right')

# Mostrar el gráfico
plt.show()
```
